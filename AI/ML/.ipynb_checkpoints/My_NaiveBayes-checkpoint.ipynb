公式：
> **B**是一个向量
$$ P(A|B) = P(B|A) * P(A) / P(B) $$

Input：
1. x：训练数据
2. y：对应的类别

Output：
1. $$P(Bi|A)$$



```python
def train_NaiveBayes(x, y, log = False):
    ResMap={}
    data_n , data_class_len = x.shape
    
    
    #统计P(C=是) P(C=否)
    if(log):
        print("统计P(C=是) P(C=否)")
    yesX = x[y=="是"]
    noX = x[y=="否"]
    if(log):
        print()
        print("yesX", len(yesX))
#         print(len(yesX))
#         print(yesX.size)
    ResMap["是"] = len(yesX)
    ResMap["否"] = len(noX)
    
    #统计C(xi)
    if(log):
        print()
        print("统计C(xi)")
    for j in range(data_class_len):
        for i in range(data_n):
            if(log):
                print(x[i][j], y[i], i)
            if(type(x[i][j]) == str):
                Key = x[i][j] + y[i]            
                if Key not in ResMap:
                    ResMap[Key]=1
                else:                
                    ResMap[Key]+=1
            else:
                break
        if(log):
            print()
    
    #计算P(xi|Ci)
    if(log):
        print("计算P(xi|Ci)")
    for key in ResMap.keys():
        if(len(key) >1):
            if(key[-1:] == "是"):
                ResMap[key] /= ResMap["是"]
            else:
                ResMap[key] /= ResMap["否"]
            print(key, ResMap[key])
        
   
    #计算 连续型数值的 贝叶斯参数 （高斯分布的均值与标准差）
    if(log):
        print()
        print("计算 连续型数值的 贝叶斯参数 （高斯分布的均值与标准差）")
    for j in range(data_class_len):
        if(type(x[i][j]) == float):
            ResMap[str(j)+"ClassMean是"] = np.mean(yesX.T[j])
            ResMap[str(j)+"ClassVar是"] =  np.std(yesX.T[j])
            ResMap[str(j)+"ClassMean否"] = np.mean(noX.T[j])
            ResMap[str(j)+"ClassVar否"] =  np.std(noX.T[j])
    
    return ResMap
```

```python
params = train_NaiveBayes(train_data, train_label, True)
params
```

```python
import math
def gaussian(x, u, sigma):
    y = np.exp(-(x - u) ** 2 / (2 * sigma ** 2)) / (sigma * math.sqrt(2 * math.pi))
    return y
```

```python
def predict(Param, data, log=False):
    P = [1.0, 1.0]# 分两类 好瓜，不好瓜
    for i in range(len(data)):
        if(type(data[i]) == str):
            if(log):
                print("P(%s|是)"% data[i],"=",Param[data[i] +"是"])
                print("P(%s|否)"% data[i],"=",Param[data[i] +"否"])
            P[0] *=  Param[data[i] +"是"]
            P[1] *=  Param[data[i] +"否"]
        else:
            yesP = gaussian(data[i], Param[str(i)+ "ClassMean是"],  Param[str(i)+ "ClassVar是"])
            noP = gaussian(data[i], Param[str(i)+ "ClassMean否"],  Param[str(i)+ "ClassVar否"])
            if(log):
                index_list = data.index.tolist()
                print("P(%s|是)"% index_list[i], yesP)
                print("P(%s|否)"% index_list[i], noP)
            P[0] *=  yesP
            P[1] *=  noP
        
    P[0] *= Param["是"]            
    P[1] *= Param["否"]
    
    if(log):
        print(P)
    if(P[0] > P[1]):
        print("是好瓜")
    else:
        print("不是好瓜")
    
        
```
