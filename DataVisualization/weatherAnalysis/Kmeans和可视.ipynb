```python
from sklearn.cluster import KMeans
from sklearn.externals import joblib
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
# decathlon.com.cn/zh/
# https://blog.csdn.net/sinat_26917383/article/details/70240628
# https://www.urlteam.org/2016/10/%E4%BD%BF%E7%94%A8python-sklearn-%E6%9C%BA%E5%99%A8%E5%AD%A6%E4%B9%A0%E6%A1%86%E6%9E%B6%E9%92%88%E5%AF%B9140w%E4%B8%AA%E7%82%B9%E8%BF%9B%E8%A1%8Ckmeans%E5%9F%BA%E4%BA%8E%E5%AF%86%E5%BA%A6%E8%81%9A/

if __name__ == '__main__':
    ## step 1: 
    print ("step 1: load data...")
    dataSet = []
    fileIn = open('data/data.txt')
    for line in fileIn.readlines():
        lineArr = line.strip().split(' ')
        dataSet.append([float(lineArr[0]), float(lineArr[1])])

    #k
    for k in range(2,10):
        clf = KMeans(n_clusters=k) #k  KMeans
        s = clf.fit(dataSet) #
        numSamples = len(dataSet) 
        centroids = clf.labels_
        print (centroids,type(centroids)) #
        print (clf.inertia_ ) #
        mark = ['or', 'ob', 'og', 'ok', '^r', '+r', 'sr', 'dr', '<r', 'pr']
        # 
        for i in range(numSamples):
            #markIndex = int(clusterAssment[i, 0])
            plt.plot(dataSet[i][0], dataSet[i][1], mark[clf.labels_[i]]) #mark[markIndex])
        mark = ['Dr', 'Db', 'Dg', 'Dk', '^b', '+b', 'sb', 'db', '<b', 'pb']
        # 
        centroids =  clf.cluster_centers_
        for i in range(k):
            plt.plot(centroids[i][0], centroids[i][1], mark[i], markersize = 12)
            #print centroids[i, 0], centroids[i, 1]
        plt.show()
```
